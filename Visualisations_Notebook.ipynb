```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import warnings

def plot_readmission_rate(df, column_name, title, boundaries=None):
    with warnings.catch_warnings():
        warnings.filterwarnings('ignore', category=pd.errors.SettingWithCopyWarning)

        # Remove rows where the column of interest or 'readmitted_next' is NaN
        df = df.dropna(subset=[column_name, 'readmitted_next'])

        if df[column_name].dtype == 'object':
            # For string or object dtype, proceed directly to grouping
            grouped_column = column_name
        else:
            # Round up the values in the specified column to the nearest integer for numerical dtype
            df[f'{column_name}_rounded'] = np.ceil(df[column_name])

            # Prepare custom bins including intervals for "lower than first" and "higher than last"
            lower = float('-inf')
            upper = float('inf')
            custom_boundaries = [lower] + boundaries + [upper]

            # Generate labels
            custom_labels = [f'<{boundaries[0]}'] + [f'{boundaries[i-1]}-{boundaries[i]}' for i in range(1, len(boundaries))] + [f'>{boundaries[-1]}']

            # Bin the data into the custom ranges
            df[f'{column_name}_binned'] = pd.cut(df[f'{column_name}_rounded'], bins=custom_boundaries, labels=custom_labels)
            grouped_column = f'{column_name}_binned'

        # Group by the relevant column and calculate the mean readmission rate
        grouped_df = df.groupby(grouped_column)['readmitted_next'].mean() * 100

        # Drop groups where there are no values (NaN readmission rate)
        grouped_df = grouped_df.dropna()

        # Plotting
        if not grouped_df.empty:
            plt.figure(figsize=(10, 6))
            sns.barplot(x=grouped_df.index.astype(str), y=grouped_df.values, palette='flare')
            plt.title(f'Readmission Rate by {title}')
            plt.xlabel(title)
            plt.ylabel('Readmission Rate (%)')
            plt.show()
        else:
            print(f"No data available for plotting Readmission Rate by {title}.")

```
